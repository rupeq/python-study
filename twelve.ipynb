```python
def create_user(login, pwd):
    if login == pwd:
        raise ValueError("pwd and login is not correct")
    flag = [False, False, False]
    for letter in pwd:
        if letter.islower():
            flag[0] = True
        if letter.isupper():
            flag[1] = True
        if letter.isdigit():
            flag[2] = True
    if all(flag):
        return "register is done"
    else:
        raise ValueError("pwd is not correct")
```

```python
while True:
    row = input("enter the cordinates")
    login, pwd = row.split()
    try:
        print(create_user(login, pwd))
    except ValueError as ve:
        print(ve.args)
    else:
        break
```

```python
from string import ascii_uppercase, ascii_lowercase, digits


ascii_uppercase, ascii_lowercase, digits = set(ascii_uppercase), set(ascii_lowercase), set(digits)

def create_user(login, pwd):
    if login == pwd:
        raise ValueError("pwd and login is not correct")
    set_pwd = set(pwd)
    if set_pwd & ascii_uppercase and set_pwd & ascii_lowercase and set_pwd & digits:
        return "Done"
    else:
        raise ValueError("pwd and login is not correct")
        
create_user("login", "l0giN")
```

```python
from doctest import testmod


def fact(x):
    '''
    >>> fact(5)
    120
    >>> fact(0)
    1
    >>> fact(-1)
    Traceback (most recent call last):
        ...
    ValueError: x must be positive integer
    >>> fact(0.5)
    Traceback (most recent call last):
        ...
    ValueError: x must be positive integer
    '''
    if x < 0 or type(x) != int:
        raise ValueError("x must be positive integer")
    return x * fact(x - 1) if x else 1


testmod()  # TDD Test Drive Dev
```

```python
roads = {
    "A":{"B":5, "C":8, "D":12, "E":76},
    "B":{"A":1, "C":7, "D":13, "E":66},
    "C":{"B":4, "A":9, "D":20, "E":49},
    "D":{"B":2, "C":5, "A":43, "E":10},
    "E":{"B":6, "C":53, "D":6, "A":60},
}
```

```python
def find_way(way):
    result = 0
    for i in range(len(way) - 1):
        start, finish = way[i:i + 2]
        result += roads[start][finish]
    return result
```

```python
find_way("ABCAD")
```

```python
import os

path = r"C:\Users\Student\books\\"
all_books = os.listdir(path)
```

```python
result = {}

for book in all_books:
    try:
        with open(path + book) as file:
            content = file.read().lower()
            content = content.replace(",", " ").replace(".", " ")
            content = content.split()
            content = set(content)
            for word in content:
                if word in result:
                    result[word].append(book)
                else:
                    result[word] = [book]
    except UnicodeDecodeError as ude:
        print(ude.args)
        print(book)
```

```python
def find_words(res, *args):
    first = set(res[args[0]])
    for word in args[1:]:
        first &= set(res[word])
    return first
```

```python
find_words(result, "карты", "деньги", "два", "ствола")
```
