```python
first_div = 3
second_div = 5

for digit in range(1, 101):
    if digit % second_div == 0 and digit % first_div == 0:
        print(f'{digit} foobar')
    elif digit % second_div == 0:
        print(f"{digit} foo")
    elif digit % first_div == 0:
        print(f"{digit} bar")
```

```python
first_div = 3
second_div = 5

for digit in range(1, 101):
    if digit % (first_div * second_div ) == 0:
        print(f'{digit} foobar')
    if digit % second_div == 0:
        print(f"{digit} foo")
    if digit % first_div == 0:
        print(f"{digit} bar")
```

```python
digit = input("enter the digit")
digit = int(digit)
length_digit = 5
count = 5
result = []

for i in range(10 ** (length_digit - 1), 10 ** length_digit):
    current_sum = 0
    origin_digit = i
    while i:
        current_sum += i % 10
        i //= 10
    if current_sum == digit:
        result.append(origin_digit)
        count -= 1
        if count == 0:
            break
print(result)
```
